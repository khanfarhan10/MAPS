```python
VALS=list(LOST.values()) 
def latitude(x):
    vals=x.split(",")
    return float(vals[0])
def longitude(x):
    vals=x.split(",")
    return float(vals[1])
LOST_LATS=list(map(latitude, VALS))
LOST_LNGS=list(map(longitude, VALS))
```
